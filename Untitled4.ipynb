```python
#Under-sampling involves reducing the number of examples in the majority class to balance it with minority class.
# Separate the minority and majority classes
df_spam = df[df['class_label'] == 'spam']
df_ham = df[df['class_label'] == 'ham']

# Under-sample the majority class
df_ham_under = df_ham.sample(len(df_spam), random_state=42)

# Combine the under-sampled majority class with the minority class
df_balanced = pd.concat([df_ham_under, df_spam])

# Shuffle the dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the distribution
print(df_balanced['class_label'].value_counts())

```

```python
#oversampling involves increase minority samples by duplicating it
# Over-sample the minority class
df_spam_over = df_spam.sample(len(df_ham), replace=True, random_state=42)

# Combine the over-sampled minority class with the majority class
df_balanced = pd.concat([df_ham, df_spam_over])

# Shuffle the dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the distribution
print(df_balanced['class_label'].value_counts())
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

# Sample data (replace with your actual DataFrame)
df = pd.DataFrame({
    'message': ['free money', 'hi, how are you?', 'win a lottery', 'meeting tomorrow',
                'call now', 'your account has been compromised', 'limited time offer', 'see you soon'],
    'class_label': ['spam', 'ham', 'spam', 'ham',
                    'spam', 'spam', 'spam', 'ham']
})

# Convert class labels to numeric
df['class_label'] = df['class_label'].apply(lambda x: 1 if x == 'spam' else 0)

# Prepare features and labels
X = df['message']
y = df['class_label']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Convert training and test data to lists
lst_train = X_train.tolist()
lst_test = X_test.tolist()

# Initialize and apply TF-IDF Vectorizer
vectorizer = TfidfVectorizer(
    lowercase=True,        # convert to lowercase before tokenizing
    stop_words='english',  # remove stop words
    ngram_range=(1, 2),    # Use unigrams and bigrams
    max_features=1000      # Limit the number of features
)

# Fit the vectorizer on the training data and transform both train and test data
features_train_transformed = vectorizer.fit_transform(lst_train)
features_test_transformed = vectorizer.transform(lst_test)

# Define classifiers and their parameter grids
classifiers = {
    'MultinomialNB': MultinomialNB(),
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'SVC': SVC(),
    'RandomForestClassifier': RandomForestClassifier()
}

parameter_grids = {
    'MultinomialNB': {'alpha': [0.1, 0.5, 1.0, 2.0], 'fit_prior': [True, False]},
    'LogisticRegression': {'C': [0.1, 1.0, 10.0], 'solver': ['liblinear', 'lbfgs']},
    'SVC': {'C': [0.1, 1.0, 10.0], 'kernel': ['linear', 'rbf']},
    'RandomForestClassifier': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
}

# Use StratifiedKFold with an appropriate number of splits
n_splits = min(len(y_train.unique()), 5)  # Ensure n_splits is not greater than the number of unique classes

# Handle very small datasets
if n_splits < 2:
    n_splits = 2

skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=0)

# Grid Search for each classifier
best_models = {}
for clf_name, clf in classifiers.items():
    print(f"Training and tuning {clf_name}...")
    grid_search = GridSearchCV(clf, parameter_grids[clf_name], cv=skf, scoring='accuracy', n_jobs=-1)
    grid_search.fit(features_train_transformed, y_train)
    best_models[clf_name] = grid_search.best_estimator_

# Evaluate the best model
for clf_name, model in best_models.items():
    y_pred = model.predict(features_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'\n{clf_name} Accuracy:', accuracy)
    print(f'{clf_name} Classification Report:\n', classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, f1_score

# Evaluate the best model
for clf_name, model in best_models.items():
    # Generate predictions
    y_pred = model.predict(features_test_transformed)
    
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Compute F1-Score
    f1 = f1_score(y_test, y_pred, average='weighted')  # 'weighted' handles imbalance
    print(f'\n{clf_name} F1-Score:', f1)
    
    # Print Classification Report
    print(f'{clf_name} Classification Report:\n', classification_report(y_test, y_pred))
    
    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f'{clf_name} Confusion Matrix')
    plt.colorbar()
    
    classes = ['Not Spam', 'Spam']
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, f'{cm[i, j]}',
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()
```

```python
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Example data (replace this with your actual training data)
x_train = ["Free entry in 2 a weekly competition to win FA Cup final tickets", 
           "Hello, can we meet tomorrow for a coffee?",
           "Congratulations! You've won a $1000 gift card."]
y_train = [1, 0, 1]  # Labels (1 = Spam, 0 = Ham)

# Initialize and fit the vectorizer
vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english'
)
features_train_transformed = vectorizer.fit_transform(x_train)

# Initialize and train the model
model = MultinomialNB()
model.fit(features_train_transformed, y_train)

# Save the vectorizer and model
joblib.dump(vectorizer, 'vectorizer.pkl')
joblib.dump(model, 'final_model.pkl')
```
